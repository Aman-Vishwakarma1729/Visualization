Matplotlib is a Python library that provides a comprehensive set of tools for creating static, animated, and interactive visualizations in Python. It is one of the most widely used data visualization libraries in Python.

Matplotlib is used to create various types of plots such as line plots, scatter plots, bar plots, histograms, and many more. It provides a wide range of customization options for plots such as changing the color, adding labels and titles, and adjusting the size and style of plot elements.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

* Line plot: A line plot is used to visualize the trend of data over time. It can be used to plot one or more lines on the same plot.

* Scatter plot: A scatter plot is used to visualize the relationship between two variables. It can be used to plot points in 2D or 3D.

* Bar plot: A bar plot is used to compare the values of different categories. It can be used to plot horizontal or vertical bars.

* Histogram: A histogram is used to visualize the distribution of a numerical variable. It can be used to plot a single variable or multiple variables on the same plot.

* Pie chart: A pie chart is used to show the proportion of different categories in a dataset. It can be used to plot a single variable or multiple variables on the same plot.

#### Answer_2

A scatter plot is used to visualize the relationship between two variables. It can be used to plot points in 2D or 3D.

```python
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel("This is X-axis")
plt.ylabel("This is Y-axis")
plt.title("This is SCATTER plot")
```

#### Answer_3

* The subplot() function in Matplotlib is used to create multiple plots in a single figure. It is a useful function when you want to compare multiple plots or show different aspects of the data in one figure.



```python
import numpy as np

line_1_x = np.array([0, 1, 2, 3, 4, 5]) 
line_1_y = np.array([0, 100, 200, 300, 400, 500])

line_2_x = np.array([0, 1, 2, 3, 4, 5])
line_2_y = np.array([50, 20, 40, 20, 60, 70])

line_3_x = np.array([0, 1, 2, 3, 4, 5])
line_3_y = np.array([10, 20, 30, 40, 50, 60])

line_4_x = np.array([0, 1, 2, 3, 4, 5])
line_4_y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0,0].plot(line_1_x,line_1_y)
axs[0,1].plot(line_2_x,line_2_y)
axs[1,0].plot(line_3_x,line_3_y)
axs[1,1].plot(line_4_x,line_4_y)
```

A box plot, also known as a box-and-whisker plot, is a type of statistical chart used to display the distribution of a dataset. It is a standardized way of displaying the distribution of data based on five number summary (minimum, 1st quartile, median, 3rd quartile, maximum).

A box plot consists of a box and whiskers, which represent different aspects of the data distribution. The box spans the 1st quartile to the 3rd quartile, with the median represented by a horizontal line inside the box. The whiskers extend from the box to the minimum and maximum values, excluding outliers. Outliers, if present, are represented as individual points outside the whiskers.

Box plots are used to identify the central tendency and spread of a dataset and to detect outliers. They are particularly useful when comparing multiple datasets, as they allow for easy comparison of the medians, quartiles, and ranges between groups.

Box plots can also be used to visualize data based on different categories or groups. For example, you can create a box plot to display the distribution of a variable by different age groups or by different regions.

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2)

box1 = np.random.normal(100, 10, 200)
axes[0].boxplot(box1, boxprops=dict(color='red'), whiskerprops=dict(color='red'))

box2 = np.random.normal(90, 20, 200)
axes[1].boxplot(box2, boxprops=dict(color='green'), whiskerprops=dict(color='green'))
```
